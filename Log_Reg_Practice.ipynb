```python
bankdf.rename(columns={'y': 'Loan_Status_Y'},inplace=True)
```

```python
bdf=bankdf.drop(['euribor3m'],axis=1)
bdf
```

```python
bankdfdum=pd.get_dummies(bdf,drop_first=True)
bankdfdum
```

```python
bankdfdum.info()
```

```python
Xtrain=bankdfdum.drop(['Loan_Status_Y'],axis=1)
Ytrain=bankdfdum['Loan_Status_Y']

Xtest=bankdfdum.drop(['Loan_Status_Y'],axis=1)
Ytest=bankdfdum['Loan_Status_Y']
```

```python
from sklearn.linear_model import LogisticRegression as LR1
```

```python
log_reg=LR1()
```

```python
log_reg.fit(Xtrain,Ytrain)
```

```python
y_predict=log_reg.predict(Xtest)
```

```python
print(y_predict)
```

```python
if len(Ytest)==len(y_predict) : print("Length compatible")
```

```python
l1=list(Ytest)
l2=list(y_predict)
```

```python
count=0
for m,n in zip(l1,l2):
        if m ==n:
            count += 1
            
print('matches=',count)
print('accuracy=',count*100/len(l1))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(Ytest, y_predict))
```
